# **Exemplo: Cramer**
Resolver o seguinte sistemas linear:
$$
\left\{
\begin{align}
& x+ y +z=4\\
&x+2y+z=5\\
&x-y-z=0
\end{align}
\right.
$$

```python
# Exemplo de Cramer
import numpy as np
#cria a matriz incompleta
matriz_D=np.matrix([[1,1,1],[1,2,1],[1,-1,-1]])
#calcula o determinante de D
D=np.linalg.det(matriz_D)
print(round(D,2))
#cria a matriz Dx
matriz_Dx=np.matrix([[4,1,1],[5,2,1],[0,-1,-1]])
Dx=np.linalg.det(matriz_Dx)
#cria a matriz Dy
matriz_Dy=np.matrix([[1,4,1],[1,5,1],[1,0,-1]])
Dy=np.linalg.det(matriz_Dy)
#cria a matriz Dz
matriz_Dz=np.matrix([[1,1,4],[1,2,5],[1,-1,0]])
Dz=np.linalg.det(matriz_Dz)
x=Dx/D
y=Dy/D
z=Dz/D
print("x=",np.round(x,2))
print("y=",np.round(y,2))
print("z=", np.round(z,2))
```

# **Exemplo 2: Eliminação de Gauss**
Resolver o seguinte sistemas linear:
$$
\left\{
\begin{align}
& x+ y +z=4\\
&x+2y+z=5\\
&x-y-z=0
\end{align}
\right.
$$

```python
# Resolução de Sistemas Lineares

# Eliminação de Gauss

import numpy as np
x=[]

def gaussElimin(C,n,m):
    for k in range (0,n):
        if C[k,k]==0:
            for s in range (k+1,n):
                if C[s,k]!=0:
                    aux=C[k,:]
                    C[k,:]=C[s,:]
                    C[s,:]=aux
        for i in range (k+1,n):
            mik=C[i,k]/C[k,k]
            for j in range (k,m):
                C[i,j]=C[i,j]-mik*C[k,j] 
                      
    return C

A=np.array([[1,1,1,4],[1,2,1,5],[1,-1,-1,0]])
#transfoma em float
A=A.astype(float)
n = len(A)
m = len(A[0])
X= gaussElimin(A,n,m)
if X[n-1,m-2]==0 and X[n-1,m-1]==0:
    print('o sistema é indeterminado')
elif X[n-1,m-2]==0 and [n-1,m-1]!=0:
    print('o sistema é impossível')
else:
    x = n*[0]
    for i in range(n-1, -1, -1):
        s = sum([A[i][j]*x[j]for j in range(i+1,n)])
        x[i] = (A[i][n] - s)/A[i][i]
    print(x)
```

# **Exemplo 4**

$$
\left\{
\begin{align}
& x+y +z=4\\
& x+2y+z=5\\
& x-y+z=0
\end{align}
\right.
$$

```python
#Função matrix para definir matrizes
#função linalg para fazer matiz inversa
import numpy as np
#cria a matriz A dos coeficientes
matriz_A=np.matrix([[1,1,1],[1,2,1],[1,-1,-1]])
#calcula o determinante de A
D=np.linalg.det(matriz_A)
print(round(D,2))
#cria a matriz B coluna
matriz_B=([[4],[5],[0]])
#cria a matriz inversa de A
matriz_A_inv=np.linalg.inv(matriz_A)
#Resolve o sistema linear
sol=matriz_A_inv*matriz_B
print("Resultado do sistema linear")
print(sol)

# Se a determinante der 0 não vai ser possivel calcular a resposta (Similar ao método de Cramer)
```

## Método iterativo de Gauss Seidel

```python
import numpy as np
# Método numérico - a ideia é supor numeros e ficar colocando no lugar das variaveis, até que a diferença do número calculado para o suposto seja pequena
# Por exemplo ele coloca em y e z o valor 1, e calcula x, com esse x ele vai para a função de calcular y e coloca o z imaginado e o x encontrado
# E repete esse processo em todos os casos

array2D = np.array([[5,1,1,1],[1,-10,-1,2],[1,1,-5,4]])

x0 = array2D[0,3]/array2D[0,0]

y0 = array2D[1,3]/array2D[1,1]

z0 = array2D[2,3]/array2D[2,2]

cont=0
while True:
  x = (array2D[0,3]/array2D[0,0]) - (array2D[0,1]/array2D[0,0]) * y0 - (array2D[0,2]/array2D[0,0]) * z0
  y = (array2D[1,3]/array2D[1,1]) - (array2D[1,0]/array2D[1,1]) * x - (array2D[1,2]/array2D[1,1]) * z0
  z = (array2D[2,3]/array2D[2,2]) - (array2D[2,0]/array2D[2,2]) * x - (array2D[2,1]/array2D[2,2]) * y
  cont=cont+1
  if abs(x-x0)<0.01 and abs(y-y0)<0.01 and abs(z-z0)<0.01 or cont>=10 :
    break
  x0 = x
  y0 = y
  z0 = z

if cont>=10:
    print("sistema não converge para uma solução pelo método de Gauss- Seidel")
else:    
    print("X = " ,round(x,2))
    print("Y = " ,round(y,2) )
    print("Z = " , round(z,2))
```
